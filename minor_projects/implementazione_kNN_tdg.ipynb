# Esercizio: Implementazione algoritmo k-nearest neighbour (kNN)

```python
# import librerie
import pandas as pd
import numpy as np
```

```python
# import database
df_raw = pd.read_csv('data/titanic.csv')
df_raw.head(5)
```

```python
# Costruzione algoritmo kNN per caso-studio di: Supervised Learning con classificazione binaria
# autput funzione: accuratezza (=numero di volte che la predizione è giusta / il numero totale di predizioni fatte)

# La logica si basa sull'utilizzo delle distanze euclidee


def kNN(X, y, k): # X = input delle feature, y = output da predire
    Y_p = np.array(np.zeros(len(y))) # creo vettore delle predizioni (ipotesi)
    for i, v in enumerate(X): # itero sulle righe v (v = vettore della matrice X che identifica un punto)
        dx = X-v # differenza tra ogni punto e il punto iterato
        norm_q = np.sum(dx**2,axis=1) # sommo le diff^2.(lungo le righe) rispetto al punto iterato (vettore norme al quadrato)
        norm = np.sqrt(norm_q) # vettore delle distanze euclidee (norme)
        indici_norm = np.argsort(norm) # indici distanze dalla più piccola alla più grande dal punto iterato
        k_neighbours = indici_norm[1:k+1] # slicing: il primo indice 0 è della distanza del punto con se stesso e va scartato
        labels = y[k_neighbours] # output nei primi vicini
        num_survived = sum(labels==1)
        num_unsurvived = sum(labels==0)
        if num_survived > num_unsurvived: # voti sui label nell'intorno del punto
            Y_p[i] = 1
        else:
            Y_p[i] = 0
        
    
    verifica = Y_p == y # confronto: array di valori booleani per verificare dove la predizione è giusta ("true values")
    tot_p = len(Y_p) # n° predizioni totali
    accuracy = np.round((sum(verifica==True)/tot_p)*100, 2) #accuratezza a 2 cifre decimali
    return(accuracy)

# Dati di input
X_train = np.array(df_raw[['Pclass', 'SibSp', 'Parch', 'Fare']]) # matrice delle feature
Y_train = np.array(df_raw["Survived"]) # vettore delle osservazioni

score = kNN(X_train, Y_train, 3)
score
```
